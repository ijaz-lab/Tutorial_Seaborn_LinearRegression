```python
#Import Statements

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Import data from GitHUB csv file

import pandas as pd

data= pd.read_csv('https://raw.githubusercontent.com/ijaz-lab/Tutorial_Seaborn_LinearRegression/main/death-rate-from-air-pollution-per-100000.csv')


```

```python
# Display the first few rows of the DataFrame
print(data.head())

# Display the column names
print(data.columns)
```

```python

# Extract the relevant columns for the regression analysis
X = data['Year'].values.reshape(-1, 1)  # Independent variable (Year)
y = data['Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: Age-standardized (Rate)'].values.reshape(-1, 1)  # Dependent variable (Death rate)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)



```

```python
# Plot the data points and the linear regression line

plt.scatter(X, y, color='purple', label='Actual Data')
plt.plot(X, y_pred, color='Green', label='Linear Regression')
plt.xlabel('Year')
plt.ylabel('Death Rate due to Air Pollution')
plt.title('Linear Regression Analysis')
plt.legend()
plt.show()
```

```python
# Create a line chart

plt.plot(X, y, label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted')
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.title('Linear Regression: Air Pollution Deaths')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a bar chart


plt.bar(X.flatten(), y.flatten())
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.title('Bar Chart: Air Pollution Deaths')
plt.grid(True)
plt.show()

```
